```python
rating_with_title = ratings.merge(books,on='ISBN')
```

```python
rating_with_title
```

```python
num_rating = rating_with_title.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns = {'Book-Rating':'num-rating'},inplace=True)
num_rating
```

```python
rating_with_title = rating_with_title.drop('ISBN', axis=1)

# .mean()['Book-Rating'].reset_index()
#avg_rating.rename(columns = {'Book-Rating':'avg-rating'},inplace=True)
#avg_rating
```

```python
avg_rating = rating_with_title.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns = {'Book-Rating':'avg-rating'},inplace=True)
avg_rating
```

```python
popular_df = num_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num-rating']>=250].sort_values('avg-rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-rating','avg-rating']]
```

```python
x = rating_with_title.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = rating_with_title[rating_with_title['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:10]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
